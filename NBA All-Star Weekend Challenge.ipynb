# What is the average weight of all players who played during this time?

```python
nba_weight = nba['WT'].mean()
nba_weight
# 228.75
```

# What team is represented the most during this time?

```python
nba_team = nba.groupby('Team').count()
nba_team.sort_values(by='WT',ascending=False).head()
# Miami Heat
```
